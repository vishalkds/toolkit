Q1. What is NumPy, and why is it widely used in Python?
   - NumPy is a library for numerical computing in Python that supports large multidimensional arrays and mathematical operations. It is widely used because it is fast, efficient, and essential for data science, machine learning, and scientific computing.

Q2. How does broadcasting work in NumPy?
   - Broadcasting allows NumPy to perform arithmetic operations on arrays of different shapes by automatically expanding the smaller array across the larger one without copying data.

Q3. What is a Pandas DataFrame?
  - A Pandas DataFrame is a two-dimensional tabular data structure with labeled rows and columns, used for storing and manipulating structured data.

Q4. Explain the use of the groupby() method in Pandas.
  - The groupby() method is used to split data into groups, apply operations like aggregation or transformation on each group, and then combine the results.

Q5. Why is Seaborn preferred for statistical visualizations?
   - Seaborn is preferred because it creates visually appealing statistical graphics easily, integrates with Pandas, and simplifies complex visualizations.

Q6. What are the differences between NumPy arrays and Python lists?
   - NumPy arrays are fixed-type, memory-efficient, and support vectorized operations, whereas Python lists are dynamic, slower, and do not support direct mathematical operations.

Q7. What is a heatmap, and when should it be used?
   - A heatmap is a graphical representation of data using colors to show values. It is used to visualize patterns, correlations, or data intensity.

Q8. What does the term “vectorized operation” mean in NumPy?
   - Vectorized operation means applying operations to entire arrays without using explicit loops, improving speed and performance.

Q9. How does Matplotlib differ from Plotly?
  - Matplotlib is a static plotting library mainly for 2D visualizations, while Plotly is an interactive plotting library designed for dynamic, web-based, and advanced visualizations.

Q10. What is the significance of hierarchical indexing in Pandas?
   - Hierarchical indexing allows multiple levels of indexing in rows or columns, making it easier to work with higher-dimensional and complex datasets.

Q11. What is the role of Seaborn’s pairplot() function?
   - The pairplot() function is used to create pairwise plots of multiple variables in a dataset, showing both distributions and relationships.

Q12. What is the purpose of the describe() function in Pandas?
   -  The describe() function provides a statistical summary of a dataset, including count, mean, standard deviation, minimum, maximum, and quartiles.

Q13. Why is handling missing data important in Pandas?
   -  Handling missing data is important to maintain accuracy, consistency, and reliability of analysis, as missing values can distort results and lead to incorrect conclusions.

Q14. What are the benefits of using Plotly for data visualization?
   - Plotly provides interactive, customizable, and web-friendly visualizations with support for 2D and 3D plots, making it suitable for dashboards and data exploration.

Q15. How does NumPy handle multidimensional arrays?
   -  NumPy handles multidimensional arrays using the ndarray object, which allows efficient storage, indexing, and mathematical operations across multiple dimensions.

Q16. What is the role of Bokeh in data visualization?
   - Bokeh is a Python library for creating interactive and web-based visualizations that can handle large datasets efficiently and provide dynamic dashboards.

Q17. Explain the difference between apply() and map() in Pandas.
   -  The apply() function can be used on both rows and columns for complex operations, while the map() function is mainly used on Series for element-wise transformations.

Q18. What are some advanced features of NumPy?
   -  Advanced features of NumPy include broadcasting, linear algebra operations, Fourier transforms, random number generation, and memory-efficient multidimensional arrays.

Q19. How does Pandas simplify time series analysis?
  -  Pandas simplifies time series analysis with features like date-time indexing, resampling, shifting, rolling statistics, and built-in time-based grouping.

Q20. What is the role of a pivot table in Pandas?
   - A pivot table summarizes and reorganizes data by grouping values across rows and columns, making it easier to analyze large datasets.

Q21. Why is NumPy’s array slicing faster than Python’s list slicing?
   - NumPy’s array slicing is faster because it uses memory-efficient views of the original data, while Python lists create new objects during slicing.

Q22. What are some common use cases for Seaborn?
   -  Common use cases for Seaborn include creating heatmaps, distribution plots, categorical plots, regression plots, and pairwise comparisons for statistical analysis.
